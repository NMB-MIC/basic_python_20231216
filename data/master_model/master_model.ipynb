## Upload excel file

```python
import pandas as pd
from datetime import date
import os
```

```python
path = os.getcwd()
file_extension = '.xls'

xlxs_file_list = []
for root,dirs,files in os.walk(path):
    for name in files: 
        if name.endswith(file_extension):
            file_path = os.path.join(root,name)
            xlxs_file_list.append(file_path)
print(xlxs_file_list)
```

```python
today = date.today()
# today_date = str(today).split("-")[-1]
# month = str(today).split("-")[-2]
# year = str(today).split("-")[-3]

dfs = []
for f in xlxs_file_list:
    df = pd.read_excel(f)
    dfs.append(df)
    
```

```python
dfs = pd.concat(dfs)
dfs
```

```python
df1 = dfs.copy()
df1['serial'] = df1['model_name']+str(today)
df1
```

```python
from sqlalchemy import create_engine
engine = create_engine('postgresql://postgres:1234@192.168.1.102:5432/test')

with engine.connect() as connection:
  df1.to_sql('test_table5', connection, index=False,if_exists='append')
```
